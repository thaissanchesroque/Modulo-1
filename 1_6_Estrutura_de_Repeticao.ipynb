# For

```python
# For Range
for contador in range (10):
  print(contador)
```

```python
# For Range + Break
for contador in range(1000):
  print(contador)
  if contador == 2:
    break
```

```python
lista = ["Teste1", "Teste2"]
```

```python
#For Simples
for val in lista:
  print(val)
```

```python
#For coom Enumerate
for i, val in enumerate(lista):
  print(i, val)
```

# For com operadores compostos

```python
lista2 = [2, 3, 4]
```

```python
for val in lista2:
  val += 1
  print(val)
```

```python
for val in lista2:
  val *= 2
  print(val)
```

# While

```python
val = 1
while (val < 6):
  val *=2
  print(val)
```

```python
# while com else
val = 10
while val < 6:
  val *=2
else:
  print('Valor de val é: ', val)
```
